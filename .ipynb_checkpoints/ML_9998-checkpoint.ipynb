## Data Preparation

#### Drop unnecessary Columns
First, I remove columns that contain only unique values("Property_id", "Unnamed: 0", "Link", "Name" columns), constant value("reservations_Payment", "Canelation", "Credit_card", "Breakfst_included" columns) or just simple text("Property_Demand", "Tax", "Type_of_room" columns). Because, they do not provide any useful data for our analysis. 

```python
# Drop columns with drop() function
my_dataset.drop(labels=["Property_id", "Unnamed: 0", "Link", "Name", "reservations_Payment", "Canelation", "Credit_card",
                        "Breakfst_included", "Property_Demand", "Tax", "Type_of_room"], axis="columns", inplace=True)
```

#### Identify duplicate values in a dataset
if we look at the code below, we can see that, it is checking for duplicate values in the dataset. For example, If the row in the dataset is duplicated, it will print "True" message, otherwise it will print nothing. According to the output below, There is no duplicate row in the dataset.

```python
a = my_dataset.duplicated()
for x in a:
    if x is True:
        print(x)
    else:
        pass
```

As we can see from the output above, there are some missing values in the dataset.We will delete all rows with missing values. For that, we use dropna() function

#### Correcting errors
As we discussed in Exploratory Data Analysis stage, "Price", "Customers_Review" columns must be converted from object data type to float64 data type. Because, these columns contain numerical data and they can provide useful data in our analysis. However, "Price" column values include not only hotel price but also Arabic currency("SAR"). So, before converting the column data type, we need to remove the word "SAR'.

```python
my_dataset[['Price', 'Customers_Review']].head()
```

```python
# First, we remove the word "SAR"(Arabic currency) from every row in "Price" column and replace "," character with "." 
# As a result, in every row, there is only numeric data and we can convert it to float data type easily.
old_data = []
new_data = []
for x in my_dataset["Price"]:
    old_data.append(x) # we are collecting list of old data for replacing it with new data
    result = x[4:]
    if "," in result:
        result = result.replace(",", ".")
    new_data.append(result) 

my_dataset["Price"].replace(to_replace=old_data, value=new_data, inplace=True)
# convert price column data type to float
my_dataset['Price'] = my_dataset['Price'].astype(float)
print(my_dataset["Price"][:5]) #result
```

The same goes for the "Customers_Review" column. First, we remove the word "reviews" from every row in the column and replace character ","  with character "." 
As a result, in every row, there is only numeric data and we can convert it to float data type easily.

```python
old_data = []
new_data = []

for x in my_dataset["Customers_Review"]:
    x = str(x)
    old_data.append(x)
    r_index = x.find("reviews") # find the index of a word "reviews"
    r_index -= 1 # There is a space before the word "reviews", which we use as the last index for getting the result.
    result = x[0:r_index]
    if "," in result:
        result = result.replace(",", ".") # replace character "," with character ","
        new_data.append(result)
    else:
        new_data.append(result)
        
my_dataset["Customers_Review"].replace(to_replace=old_data, value=new_data, inplace=True)
# convert price column data type to float
my_dataset['Customers_Review'] = my_dataset['Customers_Review'].astype(float)
print(my_dataset["Customers_Review"][:5]) #result
```

As I mentioned in Exploratory Data Analysis stage, categorical columns contain a lot of unique values (It can bee seen in the output below). As a result, it can be difficult to analyze data. So, we should drop non-duplicate unique values in categorical columns. Because, non-duplicate values(frequency is equal to 1) in categorical columns do not significantly affect our data anlysis.

```python
my_dataset.describe(include=['object'])
```

```python
# this is a function to delete non-duplicate values in categorical columns
def delete_not_duplicated_rows(column):
    global my_dataset # my_dataset is a global variable, so we declare that variable as global
    duplicated = my_dataset[column].duplicated() # Returns boolean Series denoting duplicate rows.
    # By converting the variable "duplicated"  to a list,
    # we can copy the values of the variable "repeated" 
    duplicated_values = list(my_dataset[column][duplicated]) # return list of duplicate values in the column
    
    for val in my_dataset[column]:
        if val not in duplicated_values:
            # if the value is not duplicated in column, it will be replaced with a word called "not duplicated"
            my_dataset[column].replace(to_replace=val, value="not duplicated", inplace=True)
        else:
            pass
    # dropping non-duplicate values in categorical columns
    my_dataset.drop(my_dataset[my_dataset[column] == 'not duplicated'].index, inplace=True)
#list of columns that non-duplicate values need to be deleted
for column in ["City", "Max_persons", "Bed_type", "Review_title"]: 
    delete_not_duplicated_rows(column)
```

As you can see from the output below, in categorical columns, we've manged to decrease unique values that does not significantly affect our analysis.

```python
my_dataset.describe(include=['object'])
```

Most machine learning models only accept numerical variables. Preprocessing the categorical variables becomes a necessary step. We need to convert these categorical variables to numbers such that the model is able to understand and extract valuable information. We use Ordinal Encoding technique for converting categorical columns

```python
from sklearn.preprocessing import OrdinalEncoder
categorical_columns = ["City", "Max_persons", "Bed_type", "Review_title"]
ord_enc = OrdinalEncoder()
my_dataset[categorical_columns] = ord_enc.fit_transform(my_dataset[categorical_columns])
```

#### Training and Evaluation sets
Our target column in the dataset is "Price". Because, we try to predict the price of hotels in Saudi Arabia. In the code below, train_test_split splits the data and returns a list which contains four NumPy arrays while train_size = 0.2 puts 20 percent of the data into a training set and the remaining 80 percent into a testing set.

```python
# Identifying independent and dependant variables
X = my_dataset.drop('Price',axis=1)
y = my_dataset["Price"].values

# Training and Evaulation sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Models Training
First, we'll define the regression model by using the DecisionTreeRegressor class.  Then, we'll fit the model on train data and check the model accuracy score.

```python
# import the regressor
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
# Fitting the model
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
# Accuracy score
score = regressor.score(X_train, y_train)
print("R-squared:", score) 
```

#### Evaluation Metrics
The current metrics shows the error of our model. Mean Absolute Error is the average of all absolute errors. Mean Squared Error measures the average of the squares of the errors— that is, the average squared difference between the estimated values and the actual value.

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

#### Visualizing the difference between actual and predicted Price values
It can bee seen from the output below that our model has almost predicted accurate data  

```python
from matplotlib import pyplot as plt
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, linewidth=1, label="original")
plt.plot(x_ax, y_pred, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```
